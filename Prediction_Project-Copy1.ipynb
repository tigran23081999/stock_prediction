# Forecasting exchange rates and stock prices using random processes

Nowadays, more and more people are investing in investments, which raises a problem: How to find interesting investment instruments and how to predict how they will behave in the future? To answer this question, we turn to the models from the course on random processes. To begin with, let's understand how the process describes the price of stocks, currencies. If the values ​​of a variable change unpredictably over time, then they say that it obeys a stochastic process.

Let's try to find a formula that can be used to predict the value of a stock. For this, we will introduce several definitions.

A stochastic process is a process in which time is one of the random variables.
Distinguish between stochastic processes with discrete and continuous time. The first ones take place when the values ​​of the variable change only at fixed times. In a continuous-time process, this can happen at any time.
 
There is also a division of the stochastic processes themselves into discrete and continuous. In the case of a discrete stochastic process, the variable takes values ​​from a finite (countable) set. In the continuous case - from a certain range.
 
Consider a continuous stochastic process with continuous-time describing the change in the price of a stock.

A Markov process is a stochastic process in which the future value of a variable depends only on its previous value and does not depend on all other previous values. Forecasting exchange rates and stock prices using random processes

In this case, we will start from an arbitrary point (not necessarily from 0), this does not change the essence of the process and we can use this, therefore the random variable $Z$ obeys the Wiener process if: A random variable $Z$ obeys the Markov process and $\Delta Z = \varepsilon \cdot \sqrt{t}$,
$\Delta Z $ $\mapsto$ $0$ и $\varepsilon$ $\sim$ $N(0,1)$.

By the definition of the Markov process, the quantities $\Delta Z$ are independent, therefore, for $\Delta t$ $\mapsto$ $0$ we have: $\Delta Z$ $\mapsto$ to Wiener Process $dZ$.

The generalized Wiener process is given by the equation: $dX = a\cdot dt + b\cdot dZ $ $\text{ }  $, where $a,b$ const.

If $b=0$, then equiation transforms into $dX=a \cdot dt$, откуда $X=X(0)+ a \cdot t$.

$b \cdot dZ$ will be considered as noise ,by $b$ times greater than the value of the Wiener process $dZ$.
From the property of the Wiener process, we obtain $\Delta Z$ $= a \cdot dt + b\cdot $$\varepsilon$$ \cdot \sqrt{t}$.
![Обобщённый винеровский процесс](obob_w.png "Обобщённый винеровский процесс")

The Ito stochastic process is a generalized Wiener process in which the parameters $a$ и $b$ are functions of the random variable  $x$ and time $t$.

$dx = a(x,t)\cdot dt + b(x,t)\cdot dz$ $\text{  }$ 

$S$ - stock price at time. $t$,  $\mu$ - expected rate of return
If the volatility of the stock is $0$, then $\Delta S = \mu\cdot S\cdot dt$. If $\Delta t \mapsto 0$, we get $dS = \mu \cdot S\cdot dt$.

But in life, the volatility is almost always more than $0$. Therefore, we assume that the change in profitability is constant and does not depend on the stock price, then:
$${{dS}\over{S}} = \mu\cdot dt + \sigma\cdot dz$$

Thus, we have obtained the Ito formula, which is widely used to model the stock price.

We got a formula with which we can make predictions, but let's figure it out and understand what the main coefficients in this equation are.

Let's start with the $\sigma$ ratio, which is responsible for the volatility of a financial asset. What is volatility and why is it so important for an investment analyst? In a sense, volatility is a measure of the rate at which the market price changes. Markets where prices change slowly are called low volatility, and markets where prices change quickly are called high volatility. For example, we can consider the shares of the Carnival company, which deals with cruises in the United States. Amid the pandemic, the shares of this company fell by more than $80$%
![carnival](carnival.png "Carnival")

We can say that the company was on the verge of bankruptcy, because the US government did not financially support the cruise companies due to the fact that they are legally registered in the countries of the Caribbean. But if we look, we can see that the stock has gone from $ 8 to $ 16 in a short period. This happened due to the fact that the company had large investors from Saudi Arabia. Agree a good investment, but with such success these investors might not have been, and then the company would really be on the verge of bankruptcy. When deciding to invest in this company, we will certainly look at its background, and in this case we see that such an investment has a high risk. Thus, we came to the definition of volatility. Volatility is a measure of risk. If the volatility of the stock is high, then the profitability of such an investment will be higher, but the probability of making a profit will be lower than if we invested in highly reliable government bonds, we will receive a guaranteed coupon yield.

Quantifying volatility is based on random walk and normal distribution. Take a game of ping pong, for example. The ball rolls down through the pins picket fence.
![v1](v1.png "Fig. 1")
Having bumped into a pin, it deviates to the right or the left with equal probability. After that, the ball goes to a new level, where it hits another pin. Finally, at the bottom, he falls into one of the holes. The movement of a ball through a palisade of pins is called a random walk. Once the ball hits this picket fence, no one can influence its trajectory, as well as predict this trajectory.

If you throw a sufficient number of balls, then the distribution will be described by a bell-shaped curve, as in the figure
![v2](v2.png "Fig. 2")
Most of the balls hit the center; the further the holes are from the center, the fewer balls there are. This distribution is called normal or bell-shaped.

Let's slightly change the conditions of the game by placing the vertical partitions in such a way that now, bumping into the pin and deviating to the left or right, the ball will drop to contact with the next pin, not one, but two levels. If you throw enough balls, you get the distribution shown in the figure below.
![v3](v3.png "Fig. 3")
Since the lateral movement of the balls is limited, the peak of this curve will be higher and its edges will be narrower than that of the previous stake curve.

Finally, we can put horizontal baffles so that, getting to the next level, the ball will deflect two pins to the left or right each time. And again, if you throw enough balls, you get the distribution shown in the figure below:
![v4](v4.png "Fig. 4")
This curve, which reflects a normal distribution, has a much lower peak and tails decay much more slowly than in previous curves. Let the sideways movement of the ball symbolize the upward and downward changes in the asset price, and the downward movement - the passage of time. If we assume that the price of an asset rises or falls by the dollar every day, then the distribution will be as in Fig. 3. And if we assume that the price per day rises and falls by $ 2, then the distribution will be as in Fig. 4.

If we proceed from the fact that asset price changes are random, and do not say anything about the likely direction of change, then it can be argued that the curves reflect the distribution of price values, respectively, in the moderate, low volatility and high volatility markets. In a low volatility market, prices fluctuate in a narrow range, and therefore, changes in the price of an asset are not fast. In a highly volatile market, the likelihood of price spikes is much greater and the price can fluctuate a lot.

Having defined the concept of volatility, we will introduce a formula for its calculation:
\begin{equation} \sigma = \sqrt{\frac{365 \cdot \sum\limits_{i=1}^n (x_i - m)^2 }{(n-1)\cdot n}} \end{equation}

where:

$m$ - expected value

$n$ - amount of days

$x_i$ - standard deviation of logarithmic price changes

We have obtained a formula for calculating volatility. Let's try to get a formula for calculating $\mu$ - asset return.

We will use the CAPM financial model to calculate the yield.

If all investors in the market have the same market portfolios, how do investors react to risk aversion in their investments? In a financial asset pricing model, when investors decide how much to invest in a risk-free asset and how much in a market portfolio when allocating funds, they rely on their risk preferences. Risk-averse investors may choose to invest all their savings in a risk-free asset. Investors willing to take more risk will put a significant portion of their savings, or even all of them, in a market portfolio. Investors who have already invested all their funds in a market portfolio and yet are willing to take on even more risk could achieve this by borrowing funds at a risk-free rate and investing them in the same market portfolio, following the example of everyone else. ... These assumptions are based on two additional assumptions. First, there is a risk-free asset, the expected return of which is known with absolute certainty. Second, investors can lend and borrow at a risk-free rate to optimize their allocation. While a risk-free rate loan is not particularly problematic (an individual only needs to purchase Treasury bills or Treasury bonds), obtaining loans at a risk-free rate can be much more difficult for an individual. There are versions of the CAPM model that make it possible to somewhat soften these assumptions and, nevertheless, obtain conclusions consistent with the model.

The risk of any asset for an investor is the risk added by this asset to the investor's portfolio as a whole. In the CAPM world, where all investors own a market portfolio, the risk of an individual asset for an investor is the risk that that asset adds to the market portfolio. On an intuitive level, it is clear that if an asset moves independently of the market portfolio, then this asset will not add too much risk to the market portfolio. In other words, most of the risk of a given asset is firm-specific and therefore can be diversified. On the other hand, if the value of an asset tends to rise simultaneously with an increase in the value of the portfolio, as well as a tendency to fall when the value of the market portfolio decreases, then the asset increases the risk of the portfolio. Such an asset has more market risk and less firm-specific risk. Statistically, the added risk is measured by the covariance of the asset with the market portfolio.

Covariance is measured as a percentage, so it is difficult to judge the relative risk of an investment based on its value. For this reason, we standardize the measure of risk by dividing the covariance of each asset in the market portfolio by the variance of the market portfolio. This allows you to get an indicator of risk, which is called the beta (beta) of a given asset:

$$\beta_i = \frac{\text{Covariance of asset i with market portfolio}}{\text{Market portfolio variance}}$$

Since the covariance of the market portfolio with itself is its variance, the beta of the market portfolio (as well as its average asset) is 1. Assets whose riskiness is higher than the average level (using this measure of risk) will have a beta coefficient above one, and assets that safer than average, will have beta less than one. Risk-free assets have zero beta.

The fact that each investor holds some combination of a risk-free asset and a market portfolio leads to the conclusion that the expected return on the asset is linearly dependent on the beta of the asset. In particular, the expected return on an asset can be written as a function of the risk-free rate and the beta of this asset

$$\mu = R - \sigma \cdot(m - R) $$ где:

$\mu$ - return

$R$ - freerisk rate 

$\sigma$ - asset vollatility 

$m$ - average index return

Thus, we have determined the formula for calculating the yield. Now we can test our assumptions on the calculation of the future stock price. As an example, let's calculate the forecast prices for the Russian energy company Inter RAO UES with a ticker on the IRAO.ME exchange.
Let's take government securities for a risk-free rate, the average yield on them is 7%

To calculate the profitability, let's download the data, we are interested in the stock prices for a period of time and the cost of the main index in the country in which the company's shares are traded

```python
import numpy as np 
import yfinance as yf
import datetime as dt
import pandas_datareader.data as web
import pandas as pd
```

```python
import numpy as np 
import yfinance as yf
import datetime as dt
import pandas_datareader.data as web

print('Start Year')
start_year =  int(input())
print('Start Month')
start_month = int(input())
print('Start Day')
start_day = int(input())
print('End Year')
end_year = int(input())
print('End Month')
end_month = int(input())
print('End Day')
end_day = int(input())
print('Ticker')
ticker = str(input())
print('Index')
indexx = str(input())

start = dt.datetime(start_year,start_month,start_day)
end = dt.datetime(end_year,end_month,end_day)

df = web.DataReader(ticker,'yahoo', start, end)
di = web.DataReader(indexx,'yahoo', start, end)
```

### Having downloaded the data, let's try to calculate the volatility coefficient for our company:

```python
o=[]
x=pd.DataFrame(o)
x[['Stock']] = df[['Adj Close']]
x[['Index']] = di[['Adj Close']]
x[['Return Stock']] = x[['Stock']].pct_change()
x[['Return Index']] = x[['Index']].pct_change()
s_std = np.std(x[['Return Stock']])
sigma = s_std * 252 **0.5
print(sigma)
print(x.tail())
```

#### Received the volatility coefficient $ \ beta $. Let's calculate the expected stock return for the CAPM model:

```python
T = len(x.Stock)
y = (x[0:T-1])
#print(y.tail())
d = y[['Return Stock']].sum()/(T-1)
print(d)
```

#### We have found the required values and can calculate the profitability using the CAPM model:

```python
CAPM = 7 + sigma*(d-7)
print(CAPM)
```

We now have everything to simulate the stock price using the Monte Carlo method.

Simulation of a stochastic process using the Monte Carlo method is a procedure for selecting random process values. Earlier we found that the expected return is 5.05% and the volatility is 27.8%. Therefore, in the formula ${{\textbf{dS}}\over{\textbf{S}}} = \mu*\Delta t + \sigma*\Delta z$, $\mu = 0.002$, $\sigma = 0.278$. We take the time interval for the forecast for 1 day, therefore $\Delta t = \frac{1}{365} = 0.00274$

We get formula: $$\Delta S = 0.0505*S + 0.278*S*\sqrt{0.0274}*\varepsilon $$

Напишем программу, которая будем моделировать различные сценарии акции в будущем

```python
# Импортируем нужные нам библиотеки
import pandas_datareader.data as web
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

# Получаем данные
prices = web.DataReader(ticker, 'yahoo', start, end)['Close']

# Вычислим доходности
returns = prices.pct_change()
#print(returns)
last_price = prices[-1]

# Введем количество траекторий и на сколько дней прогнозируем
print('Simulation Number')
num_simulations = int(input())
print('Forecast days')
num_days = int(input())

simulation_df = pd.DataFrame()
#print(simulation_df)

for x in range(num_simulations):
    count = 0
    daily_vol = returns.std()
    
    price_series = []
    
    price = last_price * (1 + np.random.normal(0,daily_vol)*sigma*((0.0274)**0.5)) + (np.random.normal(0,1))*CAPM/100*((0.0274)**0.5)
    price_series.append(price)
    #print(price_series)

    
    for y in range(num_days):
        if count == 251:
            break
        price = price_series[count] * (1 + np.random.normal(0,daily_vol)*sigma*((0.0274)**0.5) + (np.random.normal(0,1)*CAPM/100)*((0.0274)**0.5)) #*0.325*((0.0274)**0.5) + (CAPM/100)*((0.0274)**0.5))
        price_series.append(price)
        count += 1
    
    simulation_df[x] = price_series
print(simulation_df.iloc[-1].mean())
    
# Изобразим все траектории на диаграме
fig = plt.figure(figsize=(20, 15))
fig.suptitle('Stohastic Simulation:')
plt.plot(simulation_df)
plt.axhline(y = last_price, color = 'r', linestyle = '-')
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```

### Stock forecast based on ML

```python
import quandl
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import yfinance as yf
import datetime as dt
import pandas_datareader.data as web
```

```python
df = web.DataReader(ticker,'yahoo', start, end)
print(df.tail())
```

```python
# From all data, we take only the closing price
df.drop('High',axis=1,inplace=True)
df.drop('Close',axis=1,inplace=True)
df.drop('Low',axis=1,inplace=True)
df.drop('Open',axis=1,inplace=True)
df.drop('Volume',axis=1,inplace=True)
print(df.tail())
```

```python
# Let's introduce a variable that describes for how many days we make a forecast
forecast_out = 7 #'n=30' days
df['Prediction'] = df[['Adj Close']].shift(-forecast_out)
#print the new data set
print(df.tail())
```

```python
X = np.array(df.drop(['Prediction'],1))
X = X[:-forecast_out]
#print(X)
```

```python
y = np.array(df['Prediction'])
y = y[:-forecast_out]
#print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.48)
```

```python
#  ( Linear Regression  Model )
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr_confidence = lr.score(x_test, y_test)
print("lr confidence: ", lr_confidence)
```

```python
x_forecast = np.array(df.drop(['Prediction'],1))[-forecast_out:]
#print(x_forecast)
```

```python
# Forecast for n days
lr_prediction = lr.predict(x_forecast)
print(lr_prediction[-1])
```
